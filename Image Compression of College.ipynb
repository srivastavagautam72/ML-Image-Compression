```python
from skimage.io import imread
```

```python
img = imread('Galgotias-university.png')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

```python
plt.imshow(img)
```

```python
plt.show()
```

```python
img = img/225
```

```python
img
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
img.shape
```

```python
img_format = img.reshape(389 * 700, 3)
```

```python
img_format
```

```python
kmeans.fit(img_format)
```

```python
clusters = kmeans.predict(img_format)
```

```python
clusters
```

```python
kmeans.cluster_centers_
```

```python
kmeans.cluster_centers_[[0,0]]
```

```python
img_new = kmeans.cluster_centers_[clusters]
```

```python
new_img = img_new.reshape(389, 700, 3)
```

```python
plt.imshow(new_img)
```
